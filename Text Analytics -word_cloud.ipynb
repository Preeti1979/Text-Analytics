```python
full_text = " ".join(product_reviews_2.words('Canon_PowerShot_SD500.txt'))
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
eng_stopwords = nltk.corpus.stopwords.words('english')
```

```python
print(eng_stopwords)
```

```python
wc = WordCloud(max_words =100, stopwords = eng_stopwords + ['time', 'take', 'good', 'one'], background_color='white')
wc.generate(full_text)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (16,16))
plt.imshow(wc)
```
