```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("Apple is looking to buy UK startup for 1 billion dollar ")
doc
```

```python
# No stemming in spacy
for token in doc:
    print(token.text, '\t', token.lemma_,'\t', token.pos_, '\t', token.tag_)
```

```python
print(doc.text)
```

```python
print(doc[7].pos_)
```

```python
print(doc[7].tag_)
```

```python
print(spacy.explain(doc[7].tag_))
```

```python
print(spacy.explain(doc[0].tag_))
```

```python
for word in doc:
    print(f'{word.text:{12}}{word.pos_:{10}}{word.tag_:{10}}{spacy.explain(word.tag_)}')
```
